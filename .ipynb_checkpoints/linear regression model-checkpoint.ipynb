```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

## Generating the data

```python
N = 1000
def function(x):
    return 5*x - 3

x = np.linspace(0,100,N)

y = np.zeros(N)
for i in range(N):
    y[i] = function(x[i])

train_input = np.zeros(500)
train_output = np.zeros(500)
test_input = np.zeros(500)
test_output = np.zeros(500)
for i in range(500):
    train_input[i] = x[i]
    train_output[i] = y[i]
    test_input[500-i-1] = x[N-i-1]
    test_output[500-i-1] = y[N-i-1]

print(test_input[0], test_output[0])
```

## Setting up the computational graph

```python
data = tf.placeholder(tf.float32, [1])
target = tf.placeholder(tf.float32, [1])

w = tf.Variable(tf.truncated_normal([1]), dtype = tf.float32)
b = tf.Variable(tf.zeros([1]), dtype = tf.float32)

myval = np.dot(data, w)

loss = tf.sqrt(np.abs((tf.add(myval, b))-((target))))
my_optimizer = tf.train.GradientDescentOptimizer(0.005)
train_step = my_optimizer.minimize(loss)
```

## Running the session

```python
init_op = tf.global_variables_initializer()

sess = tf.Session()
sess.run(init_op)
epoch = 10000
w_v = []
for i in range(epoch):
    rand_index = np.random.choice(500)
    xrand = [train_input[rand_index]]
    yrand = [train_output[rand_index]]
    sess.run(train_step, feed_dict = {data: xrand, target : yrand})
    #print(sess.run(w), sess.run(b))
    w_v.append(sess.run(w))
```

```python
figsize(12, 9)
plot(test_input, test_input*sess.run(w)+sess.run(b), test_input, test_output)
show()
```
