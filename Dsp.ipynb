# Aggregációs feladat
Aggregációs függvény: $A_y = \sum_n y[n]$

$A_y = A_h\cdot A_x$, ha $y[n] = h[n]*x[n]$.

```python
from scipy.signal import convolve

n = array([0,1, 2, 3, 4])
dt = 1/100
x = sin(2*pi*n*dt)+ 0.5*rand(5)
h = array([1,1,1, 1, 1])
```

```python
def conv(x, a):
    y=zeros(2*len(x)-1)
    h = zeros(2*len(x)-1)
    for i in range(len(x)):
        h[i]=a[i]
    for n in range(2*len(x)-1):
        for k in range(len(x)):
            y[n]= y[n]+x[k]*h[n-k]
    return y
```

```python
def Agg(x, n):
    sum = 0
    for i in range(len(n)):
        sum = sum + x[i]
    return sum
```

```python
markerline, stemlines, baseline = stem(n,x,linefmt='b')
show()
```

```python
y = convolve(x, h)
q = conv(x, h)
print(y)
print(q)
print(Agg(y,y))

print(Agg(x,x)*Agg(h,h))
```

```python

x = linspace(0,3, 10)
h = linspace(-1, 1, 10)
h = h/h
x= x/3

q = conv(x,h)
w = linspace(-2,10, len(q) )
plot(w, q)
markerline, stemlines, baseline = stem(w,q,linefmt='b')
show()
```

# Konvoluciós mátrix

```python
from scipy.linalg import *
n = array([1, 1/2, 1/4, 1/8, 1/16])
h = array([1/3, 1/9, 1/27])

def matconv(x, h):
    padding = np.zeros(h.shape[0] - 1, h.dtype)
    first_col = np.r_[x, padding]
    first_row = np.r_[x[0], padding]
    H = toeplitz(first_col, first_row)
    return H

xn = (matconv(n, h))
k = matmul(xn, h)
print(matconv(n,h))
print(k)
print(shape(h), shape(xn))
plot(k)
```

Az eredmény megegyezik a *numpy* konvoluciós függvény által adottal:

```python
q = convolve(n, h)
plot(q)
```

# Korrelációs függvény

Korellációt egyszerű számolni, mert használhatjuk a konvoluciós függvényt, csak meg kell fordítani a súlyfüggvényt.

```python
def cor(n, h):
    h = flipud(h)
    return matmul(matconv(n,h),h)
print(cor(n, h))
title("általam csinált")
plot(cor(n, h))
show()
title("beépített")
plot(correlate(n, h, mode = 'full'))
show()
```

# Egyszerű szűrő

```python
x = linspace(-10, 10, 200)
y = linspace(-10,10, 200)*200
felul = (heaviside(x-3, 1)-heaviside(x+3, 1)+1)
alul = -felul+1
szuro = alul
plot(x, szuro)
f = 3
s = sin(f*y)+sin(2*f*y)
title("A jel és a szűrő")
plot(x,s)
show()
it = real(fft.fft(s))
title("A jel fourier transzformáltja")
plot(y, abs(it))
show()
```

A szűrt jel, még frekvenciatérben:

```python
szurt = it*szuro
plot(x, szurt)
```

A jel visszatranszformálva:

```python
v = fft.ifft(szurt)
plot(x, v)
show()
```

Ilyen volt a bejövő:

```python
plot(x, s)
show()
```

```python
figsize(12,9)
title(" Az eredeti jel és a szűrt fft-je")
plot(y, fft.fft(v), y, fft.fft(s), ".")
show()
```

# spectrogram készítő függvény
A spectrogram mátrix:
$$ X[k, n] = \sum_m^{L-1}w[m]x[n+m]e^{-i(2\pi k/N)m}$$

```python
import scipy.signal
def spectrogram0(x, L, NFFT,step, Fs):
    N = len(x)
    K = int((N-L+step)/step)
    w = hanning(L)
    time = range(0,L)
    N2 = int(NFFT/2+1)
    S = zeros((K, N2))
    for k in range(1,K):
        for i in range(L):
            xw = x[time[i]]*w
            X = fft.fft(xw, NFFT)
            X1 = matrix.getH(X[1:N2])
            S[k, 1:N2] = X1*conj(X1)
            i = i + step
    S = matrix.getH(fliplr(S))
    S = S/abs(S).max()
    
    imshow(-20*log10(S))
    show()
    plot(S, ".")

x = linspace(0, 20, 500)
f1 = 100
fo = 2
t1 = 3
f0=0
sig = sin(f0 + (f1 - f0) * x / t1)
plot(x, sig)
show()
spectrogram0(sig, 20, 256, 2, 500)
```
